```python
import pickle

def save_obj(obj, name ):
    with open(name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_obj(name ):
    with open(name + '.pkl', 'rb') as f:
        return pickle.load(f)
```

```python
X_train,y_train=np.eye(256),np.eye(256)
```

```python
p=0.2
add_name='_02'
```

```python
autoencoderBAC,encoderBAC,decoderBAC=OHAutoencoder(5*256,5*256,BAC,p,use_BN=False,use_LN=False)
```

```python
boundaries = [2**15,2**16,2**17]
values = [1e-4,1e-5,1e-6,1e-7]
lr_schedule = tf.keras.optimizers.schedules.PiecewiseConstantDecay(boundaries, values)
autoencoderBAC,encoderBAC,decoderBAC=OHAutoencoder(5*256,5*256,BAC,p,use_BN=True,use_LN=False,lr=lr_schedule)
```

```python
checkpoint=tf.keras.callbacks.ModelCheckpoint('BACOHAutoencoder'+add_name+'.h5',monitor='loss',save_best_only=True)
history=autoencoderBAC.fit(x=X_train,y=y_train,batch_size=256,epochs=2**18,callbacks=[checkpoint],verbose=2)
```

```python
save_obj(history.history,'BAChistory'+add_name)
encoderBAC.save('BACOHEncoder'+add_name+'.h5')
decoderBAC.save('BACOHDecoder'+add_name+'.h5')
```
